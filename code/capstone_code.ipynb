```python
img = []
for idx in range(535):
    img.append(cv2.imread('C:/users/user/capstone/img/Plot-%04d.png' % idx, cv2.IMREAD_GRAYSCALE))
    
img = np.array(img)
X_train, X_test, y_train, y_test = train_test_split(img, np.array(label_data['emotion']), test_size=0.2, random_state=123)

X_train = X_train.astype('float32') / 255.
X_test = X_test.astype('float32') / 255.

y_train = pd.get_dummies(y_train)
y_train = np.array(y_train)
y_test = pd.get_dummies(y_test)
y_test = np.array(y_test)
```

```python
input_layer = Input((216,864))

x = Conv1D(filters = 32, kernel_size = 3, strides = 1, padding = 'same')(input_layer)
x = BatchNormalization()(x)
x = LeakyReLU()(x)


x = Conv1D(filters = 32, kernel_size = 3, strides = 2, padding = 'same')(x)
x = BatchNormalization()(x)
x = LeakyReLU()(x)


x = Conv1D(filters = 64, kernel_size = 3, strides = 1, padding = 'same')(x)
x = BatchNormalization()(x)
x = LeakyReLU()(x)


x = Conv1D(filters = 64, kernel_size = 3, strides = 2, padding = 'same')(x)
x = BatchNormalization()(x)
x = LeakyReLU()(x)


x = Flatten()(x)

x = Dense(128)(x)
x = BatchNormalization()(x)
x = LeakyReLU()(x)
x = Dropout(rate = 0.5)(x)

x = Dense(7)(x)
output_layer = Activation('softmax')(x)

model = Model(input_layer, output_layer)
```

```python
model.summary()
```

```python
optimizer = Adam(lr=0.0005)
model.compile(loss = 'categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
model.fit(X_train,
         y_train,
         batch_size=32,
         epochs=10,
         shuffle=False)
```

```python
# 결과는 [loss, accuracy] 값으로 반환 됨
model.evaluate(X_test, y_test)
```

```python
CLASSES = np.array(['ANXIETY/FEAR','DISGUST','HAPPINESS','BOREDOM','NEUTRAL VERSION','SADNESS','ANGER'])

pred = model.predict(X_test)
pred_single = CLASSES[np.argmax(pred, axis=-1)]
actual_single = CLASSES[np.argmax(y_test, axis=-1)]
```
