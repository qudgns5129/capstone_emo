# 시각화 해보기

```python
sig, sr = librosa.load('c:/users/admin/capstone_emo/data/wav/03a01Fa.wav', sr=16000)

plt.figure()
librosa.display.waveplot(sig, sr, alpha=0.5)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```

- sig : 음성의 파형 데이터
- sr : sampling rate(주파수 분석 및 파형의 시간 간격을 결정)

```python
# Fourier -> Spectrum

fft = np.fft.fft(sig)

magnitude = np.abs(fft) 

f = np.linspace(0,sr,len(magnitude))

left_spectrum = magnitude[:int(len(magnitude) / 2)]
left_f = f[:int(len(magnitude) / 2)]

plt.figure()
plt.plot(left_f, left_spectrum)
plt.xlabel("Frequency[Hz]")
plt.ylabel("Magnitude[au]")
plt.title("Power spectrum")
```

- FFT(고속 푸리에 변환) : 시간영역의 데이터를 이산화된 주파수 영역으로 바꿔주는 작업
- magnitude : 소리 크기의 수치
- f 부분 해석 : 0부터 sampling rate까지의 데이터를 len(magnitude)의 개수로 균일하게 분할해라.

```python
# STFT -> spectrogram

hop_length = 256
n_fft = 513

hop_length_duration = float(hop_length) / sr
n_fft_duration = float(n_fft) / sr

stft = librosa.stft(sig, n_fft=n_fft, hop_length=hop_length)

magnitude = np.abs(stft)

log_spectrogram = librosa.amplitude_to_db(magnitude)

plt.figure()
librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram (dB)")
```

- hop_length : 음성의 magnitude를 얼만큼 겹친 상태로 잘라서 칸으로 보여줄 것인가?
- n_fft : 음성의 길이를 얼마만큼으로 자를 것인가? (window라고 부름)
- n_fft_duration : frame length라고 하며 np.float(n_fft)/sr = 1024/22050 = 0.46... 이므로 46ms가 나온다.  
- 즉, 1개의 window의 크기를 46ms로 잡고, 이것을 window_length라고도 부른다

예를 들어 우리가 보유한 데이터가 7초이고, window_length(n_fft_duration)를 0.025초, frame_stride(window_length - hop_length)를 0.010초(10ms)단위로 뽑는다고 가정하면, 1칸은 0.015초(15ms)가 겹치도록 하여 총 700칸을 얻을 수 있다.

# 데이터를 이미지로 변환하여 저장

```python
files = glob('../data/wav/*.wav')
```

```python
#
def findDuration(fname):
    with contextlib.closing(wave.open(fname,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        sw   = f.getsampwidth()
        chan = f.getnchannels()
        duration = frames / float(rate)
        #print("File:", fname, "--->",frames, rate, sw, chan)
        return duration

def graph_spectrogram(wav_file, sr=16000):
    sig, sr = librosa.load(wav_file, sr=sr)
    n_fft = 512
    win_length = 400
    hop_length = 160
    n_mels = 80

    D = np.abs(librosa.stft(sig, n_fft=n_fft, win_length = win_length, hop_length=hop_length))
    mel_spec = librosa.feature.melspectrogram(S=D, sr=sr, n_mels=n_mels, hop_length=hop_length, win_length=win_length)
    
    fig,ax = plt.subplots(1)
    fig.subplots_adjust(left=0,right=1,bottom=0,top=1)
    ax.axis('off')
    librosa.display.specshow(librosa.amplitude_to_db(mel_spec, ref=0.00002), sr=sr, hop_length = hop_length, cmap = plt.cm.jet)
    ax.axis('off')
    plt.rcParams['figure.figsize'] = [0.75,0.5]
    fig.canvas.draw()
    size_inches  = fig.get_size_inches()
    dpi          = fig.get_dpi()
    width, height = fig.get_size_inches() * fig.get_dpi()
    
    mplimage = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
    imarray = np.reshape(mplimage, (int(height), int(width), 3))
    re_imarray = cv2.resize(imarray, (54, 36), interpolation=cv2.INTER_AREA)
    plt.close(fig)
    return re_imarray
```

# 라벨링 작업

```python
import os

raw_label = []

for idx, name in enumerate(files):
    raw_label.append(os.path.splitext(name.split('\\')[1])[0])
```

```python
speaker = []
text = []
emotion = []
version = []

for i in raw_label:
    speaker.append(i[0:2])
    text.append(i[2:5])
    emotion.append(i[5:6])
    version.append(i[6:7])
    
dicts = {'speaker':speaker,'text':text,'emotion':emotion,'version':version}
label_data = pd.DataFrame(dicts)
```

```python
y_result = np.array(label_data['emotion'])
```

# 2D CNN 해보기

```python
# Train data 80% : test data 20%로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(img_arr, y_result, test_size=0.2, shuffle=False)

X_train = X_train.astype('float32') / 255.
X_test = X_test.astype('float32') / 255.

y_train = pd.get_dummies(y_train)
y_test = pd.get_dummies(y_test)
```

```python
X_train.shape
```

```python
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(36,54,3)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(7, activation='softmax'))
model.summary()

opt = Adam(lr=0.0005)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
callbacks = [tensorflow.keras.callbacks.EarlyStopping(monitor='val_loss', patience=8)]
history = model.fit(X_train, y_train, batch_size=32, epochs=64, verbose=1, validation_split=0.2, callbacks=callbacks)
```

```python
loss_and_metrics = model.evaluate(X_test, y_test, batch_size=64)
print('loss_and_metrics : ' + str(loss_and_metrics))
```

```python
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))
plt.rcParams["figure.figsize"] = (14,4)
plt.plot(x_len, y_vloss, marker='.', c='red', label="Validation-set Loss")
plt.plot(x_len, y_loss, marker='.', c='blue', label="Train-set Loss")

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
y_vacc = history.history['val_acc']
y_acc = history.history['acc']

x_len = np.arange(len(y_acc))
plt.rcParams["figure.figsize"] = (14,4)
plt.plot(x_len, y_vacc, marker='.', c='red', label="Validation-set Accuracy")
plt.plot(x_len, y_acc, marker='.', c='blue', label="Train-set Accuracy")

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
CLASSES = np.array(['ANXIETY/FEAR','DISGUST','HAPPINESS','BOREDOM','NEUTRAL VERSION','SADNESS','ANGER'])

pred = model.predict(X_test)
pred_single = CLASSES[np.argmax(pred, axis=-1)]
actual_single = CLASSES[np.argmax(y_test, axis=-1)]
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(actual_single, pred_single)
```
